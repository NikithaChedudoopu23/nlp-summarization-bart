```python
#Load the dataset
from datasets import load_dataset

dataset = load_dataset("cnn_dailymail", "3.0.0")
```

```python
print(f"Training set size: {len(dataset['train'])}")
print(f"Validation set size: {len(dataset['validation'])}")
print(f"Test set size: {len(dataset['test'])}")
```

```python
from transformers import BartTokenizer
tokenizer = BartTokenizer.from_pretrained("facebook/bart-large-cnn")
```

```python
subset_size = 5000
val_size = 1000
test_size = 1000

dataset["train"] = dataset["train"].select(range(subset_size))
dataset["validation"] = dataset["validation"].select(range(val_size))
dataset["test"] = dataset["test"].select(range(test_size))
```

```python
# Step 3: Preprocessing
def preprocess(example):
    model_inputs = tokenizer(
        example["article"], max_length=1024, truncation=True, padding="max_length"
    )
    labels = tokenizer(
        example["highlights"], max_length=128, truncation=True, padding="max_length"
    )
    labels["input_ids"] = [
        [(label if label != tokenizer.pad_token_id else -100) for label in l]
        for l in labels["input_ids"]
    ]
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
```

```python
# Apply preprocessing to subset only
tokenized_dataset = dataset.map(preprocess, batched=True)
tokenized_subset = {
    "train": tokenized_dataset["train"].select(range(subset_size)),
    "validation": tokenized_dataset["validation"].select(range(val_size))
}
```

```python
# Step 4: Load Model and Trainer
from transformers import BartForConditionalGeneration, TrainingArguments, Trainer, DataCollatorForSeq2Seq, EarlyStoppingCallback

model = BartForConditionalGeneration.from_pretrained("facebook/bart-large-cnn")
```

```python
# freeze fewer layers
for param in model.model.encoder.layers[:3]:
    param.requires_grad = False

# Data collator
data_collator = DataCollatorForSeq2Seq(tokenizer=tokenizer, model=model)
```

```python

from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./results",
    overwrite_output_dir=True,
    learning_rate=2e-5,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    num_train_epochs=2,
    weight_decay=0.01,
    logging_dir="./logs",
    save_total_limit=1,
    report_to="none",
    load_best_model_at_end=True,
    metric_for_best_model="eval_loss",
    greater_is_better=False,
    save_strategy="epoch",
    eval_strategy="epoch"
)


trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_subset["train"],
    eval_dataset=tokenized_subset["validation"],
    tokenizer=tokenizer,
    data_collator=data_collator,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=2)]
)
```

```python
# Step 5: Train the Model
trainer.train()
```

```python
from transformers import BartForConditionalGeneration, BartTokenizer
import torch

# Load fine-tuned model
model = BartForConditionalGeneration.from_pretrained("./bart_finetuned_local", local_files_only=True)
tokenizer = BartTokenizer.from_pretrained("./bart_finetuned_local", local_files_only=True)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Function to generate summary
def generate_summary(model, tokenizer, article):
    inputs = tokenizer(article, return_tensors="pt", truncation=True, max_length=1024).to(device)
    summary_ids = model.generate(
        **inputs,
        max_length=128,
        min_length=40,
        num_beams=4,
        early_stopping=True,
        length_penalty=2.0,
        no_repeat_ngram_size=3
    )
    return tokenizer.decode(summary_ids[0], skip_special_tokens=True, clean_up_tokenization_spaces=True)


```

```python
#Testset implementation
test_dataset = dataset["test"].select(range(1000))

test_articles = [x['article'] for x in test_dataset]
test_references = [x['highlights'] for x in test_dataset]
```

```python
import torch
from tqdm import tqdm

def batch_generate_summaries(model, tokenizer, articles, batch_size=8, max_length=128):
    model.eval()
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    summaries = []

    for i in tqdm(range(0, len(articles), batch_size)):
        batch = articles[i:i + batch_size]

        # Tokenize the batch
        inputs = tokenizer(
            batch,
            return_tensors="pt",
            padding=True,
            truncation=True,
            max_length=1024
        ).to(device)

        # Generate summaries
        with torch.no_grad():
            summary_ids = model.generate(
                input_ids=inputs["input_ids"],
                attention_mask=inputs["attention_mask"],
                max_length=max_length,
                min_length=40,
                num_beams=4,
                early_stopping=True,
                length_penalty=2.0,
                no_repeat_ngram_size=3
            )

        # Decode summaries and add to list
        batch_summaries = tokenizer.batch_decode(summary_ids, skip_special_tokens=True, clean_up_tokenization_spaces=True)
        summaries.extend(batch_summaries)

    return summaries
```

```python
# Pre-trained BART summaries (batched)
baseline_summaries = batch_generate_summaries(baseline_model, baseline_tokenizer, test_articles, batch_size=8)

# Fine-tuned BART summaries (batched)
finetuned_summaries = batch_generate_summaries(model, tokenizer, test_articles, batch_size=8)
```

```python
#Quantitative metrics
import evaluate

rouge = evaluate.load("rouge")
bertscore = evaluate.load("bertscore")

# ROUGE
rouge_baseline = rouge.compute(predictions=baseline_summaries, references=test_references)
rouge_finetuned = rouge.compute(predictions=finetuned_summaries, references=test_references)

# BERTScore
bert_baseline = bertscore.compute(predictions=baseline_summaries, references=test_references, lang="en")
bert_finetuned = bertscore.compute(predictions=finetuned_summaries, references=test_references, lang="en")
```

```python
from tabulate import tabulate

comparison_table = [
    ["ROUGE-1", rouge_baseline["rouge1"], rouge_finetuned["rouge1"]],
    ["ROUGE-2", rouge_baseline["rouge2"], rouge_finetuned["rouge2"]],
    ["ROUGE-L", rouge_baseline["rougeL"], rouge_finetuned["rougeL"]],
    ["BERTScore (F1)", bert_baseline["f1"][0], bert_finetuned["f1"][0]]
]

print(tabulate(comparison_table, headers=["Metric", "Pre-trained", "Fine-tuned"], tablefmt="fancy_grid"))
```

**Qualitative Evaluation Summary**

For three randomly selected test samples, we observed the following:

1. Fluency:

Both models generated fluent and grammatically correct summaries. The fine-tuned model occasionally used more natural phrasing and domain-specific vocabulary.

2. Faithfulness:

Neither model hallucinated major facts. However, the fine-tuned model sometimes included slightly more grounded context, likely due to training on the CNN/DailyMail dataset.

3. Coverage:

The fine-tuned summaries generally covered more points from the article, while the pre-trained ones were sometimes too brief.

4. Conciseness & Redundancy:

Both models maintained concise summaries. In some cases, the pre-trained model repeated parts of the original text more literally.
